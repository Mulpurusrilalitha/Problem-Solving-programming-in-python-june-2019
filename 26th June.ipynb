```python
# Function to find the second largest number in a list

def secondLargest(li):
    # Convert the list into a unique list
    # O(n)
    unique = []
    for n in li:
        if n not in unique:
            unique.append(n)
    
    unique = sorted(unique, reverse=True)
    if len(unique) > 1:
        return unique[1]
    return -1

secondLargest([1])

# def fifthLeast(li)

# def kLargest(li, k)

def kSmallest(li, k):
    # Extract Unique elements in the list
    unique = []
    for i in li:
        if i not in unique:
            unique.append(i)
    
    # Sort the Unique list in ascending order
    unique = sorted(unique)
    
    #Check if length of unique list is greater than k
    if len(unique) >= k:
        return unique[k-1]
    return -1

kSmallest([], 1)
```

```python
# Function to identify the element with highest frequency in a list
# If many elements have the have the highest frequency, return the smallest
# highestFrequencyElement([1,2,3,9,8,7,3,4,2,1]) -> 1
def highestFrequencyElement(li):
    #Extracting unique elements with frequencies
    unique = {}
    for n in li:
        if n in unique:
            unique[n] += 1
        else:
            unique[n] = 1
    # unique = {1:2, 2:2, 3:2, 9:1, 8:1, 7:1, 4:1}
    maxfreq = max(unique.values())
    
    #Extract all keys with maxfreq in a list
    maxfreqkeys = []
    for item in unique.items():
        #item -> (key, value)
        # item[1]
        if item[1] == maxfreq:
            maxfreqkeys.append(item[0])
    
    return min(maxfreqkeys)

highestFrequencyElement([1,2,3,9,8,7,3,4,2,1])
```

Input format

, denoting the number of test cases. The first line of each test case consists of a single string made of lower case English alphabets. The second line of each test case consists of a single integer denoting

Output format

```python
# Largest number in a list
# Second Largest number in a list
# Kth largest number in a list
# Element with highest frequency
# Second Highest frequency
# Kth Highest frequency


# {a:4, g:9, i:6, p:213, c:6}
# [4,6,6,9,213]
# [213, 10, 9, 6, 6, 4]
# [a,c,g,i,p]
#k = 5
# li = []
# for item in d.items():
#    if item[1] == 6:
#        li.append(item[0])
#li = [i, c]

def kLargestFrequency(s, k):
    # Construct the frequency dictionary for all unique characters
    #unique = []
    freq = {}
    for i in s:
        if i not in freq.keys():
            freq[i] = s.count(i)
            
    # Extract unique frequencies in descending order
    values = sorted(freq.values(), reverse=True)
    uniqueValues = list(set(values))
    uniqueValues = sorted(uniqueValues, reverse=True)
    
    # Identify Kth largest frequency
    if k < len(uniqueValues):
        kvalue = uniqueValues[k-1]
    else:
        return -1

    # Get all elements with Kth largest frequency
    li = []
    for item in freq.items():
        if item[1] == kvalue:
            li.append(item[0])
    # Minimum of Kth Largest frequency
    return min(li) 

with open('../input.txt', 'r') as f:
    t= int(f.readline())
    for i in range(t):
        s = f.readline()
        k = int(f.readline())
        print(kLargestFrequency(s,k))
```
